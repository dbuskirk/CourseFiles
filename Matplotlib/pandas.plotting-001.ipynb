```python
path = "../data/"
```

```python
dfIris = pd.read_csv(path+"iris.csv")
```

```python
fig, ax = plt.subplots(figsize=(12,12))
pd.plotting.scatter_matrix(dfIris, alpha=1, ax=ax)
```

Andrews curves are sensitive to the order in which the columns are presented. The following example is commonly used in blogs, but it in fact inappropriate. Andrews curves are more appropriate for view the results of a principal component analysis (PCA), and the eigenvectors should be presented in decreasing order of explained variance.

```python
pd.plotting.andrews_curves(dfIris, 'Species')
```

```python
pd.plotting.radviz(dfIris, 'Species')
```

```python
pd.plotting.radviz(dfIris, 'Species', color= ('red', 'gold','blue'))
```

```python
pd.plotting.parallel_coordinates(

    dfIris, 'Species', color=('#556270', '#4ECDC4', '#C7F464')

)
```
