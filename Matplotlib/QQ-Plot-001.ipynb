```python
import scipy
from numpy.random import seed
from numpy.random import randn
from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot
```

```python
# seed the random number generator
seed(1)
# generate univariate observations
data = 5 * randn(100) + 50
# q-q plot
qqplot(data, line='s')
pyplot.show()
```

```python

seed(1)

data = 5 * randn(100) + 50
# q-q plot
# comparison to normal distribution is the default
# we can specify that explicitly if we like
qqplot(data, dist=scipy.stats.distributions.norm, line='s')
pyplot.show()

```

```python
# seed the random number generator
seed(1)
# generate univariate observations
data = 5 * randn(100) + 50
# q-q plot
# we can ompare to some other distribution. There are a who bunch of 'em.
qqplot(data, dist=scipy.stats.distributions.expon, line='s')
pyplot.show()
```

Here are the options for the reference line:


    ‘45’ - 45-degree line

    ‘s’ - standardized line, the expected order statistics are scaled by the standard deviation of the given sample and have the mean added to them

    ‘r’ - A regression line is fit

    ‘q’ - A line is fit through the quartiles.

    None - by default no reference line is added to the plot.
