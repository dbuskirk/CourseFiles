## Creating Subplots

```python
fig, ax = plt.subplots(1)
plt.show()
```

We can set values to optional properties when we create the subplots.

```python
fig, ax = plt.subplots(1, figsize=(8,8))
plt.show()
```

Or, we can use the references we have called fig and ax to set properties later.

```python
fig, ax = plt.subplots(1 )
fig.figsize=(4,4)
plt.show()
```

As is always the case, there are many ways to accomplish the task of creating and using subplots. Subplots can be created at once and then referenced individually, or subplots can be added one-by-one to an existing plot.<br> 

### Creating Subplots All-at-Once

```python
fig, ax = plt.subplots(4,4)

plt.show()
```

Looks pretty ugly, so we will try it again using <b>constrained_layout=True</b>.

```python
fig, ax = plt.subplots(4,4, constrained_layout=True)
```

Actually, constrained_layout=True might be a good candidate for setting as a default for this document.

```python
# mpl.rcParams.keys()
```

### fig and ax

"fig" and "ax" are far-and-away the most commonly chosen variable names to refer to figure objects and axis arrays. However, there are several different coding styles to choose from. One option is to let ax remain an array and use matrix notation when we want to access a particular axis. 

```python
fig, ax = plt.subplots(4,4)
ax[1,2].set_facecolor('red')
plt.show()
```

An alternative syntax is to capture individual axis references when we create them.

```python
fig, [[ax1, ax2],[ax3, ax4]] = plt.subplots(2,2)
ax3.set_facecolor('green')
plt.show()
```

### Creating Subplots One-At-A-Time

Some folks prefer to use the RCI notation, in which the 221 just looks like an integer, but really means we are creating an array with 2 rows, 2 columns, and the subplot we are creating is instance 1 (of a possible 4).

```python
ax1=plt.subplot(221)
ax2=plt.subplot(223, projection='polar')
```

This is exactly equivalent to 

```python
ax1=plt.subplot(2,2,1)
ax2=plt.subplot(2,2,3, projection='polar')
```

See what happens when we change the instance number of the polar plot.

```python
ax1=plt.subplot(221)
ax2=plt.subplot(222, projection='polar')
```

### Creating Subplots One-At-A-Time

#### Using figure.add_subplot( ) 

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
```

```python
fig = plt.figure()
fig.add_subplot(1, 2, 1)   #top and bottom left
#plt.plot(names, values)
fig.add_subplot(2, 2, 2)   #top right
plt.plot(names, values)
fig.add_subplot(2, 2, 4)   #bottom right 
#plt.plot(names, values)
plt.show()
```
