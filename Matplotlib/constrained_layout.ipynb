```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.gridspec as gridspec
import numpy as np

```

```python
def example_plot(ax, fontsize=12, hide_labels=False):
    ax.plot([1, 2])

    ax.locator_params(nbins=3)
    if hide_labels:
        ax.set_xticklabels([])
        ax.set_yticklabels([])
    else:
        ax.set_xlabel('x-label', fontsize=fontsize)
        ax.set_ylabel('y-label', fontsize=fontsize)
        ax.set_title('Title', fontsize=fontsize)


```

```python
plt.rcParams['savefig.facecolor'] = "0.8"
plt.rcParams['figure.figsize'] = 4.5, 4.
plt.rcParams['figure.max_open_warning'] = 50
```

No difference is observed in a single plot in Jupyter, however the difference is immediately apparent in Qt.

```python
fig, ax = plt.subplots(constrained_layout=False)
example_plot(ax, fontsize=24)
```

```python
fig, ax = plt.subplots(constrained_layout=True)
example_plot(ax, fontsize=24)
```

In Jupyter, the axis label may be overlaid by the title of the plot on the next row.

```python
# constrained_layout = False is the default
fig, axs = plt.subplots(2, 2)
for ax in axs.flat:
    example_plot(ax)
```

```python
fig, axs = plt.subplots(2, 2, constrained_layout=False)
for ax in axs.flat:
    example_plot(ax)
```

Setting constrained_layout=True fixes the problem.

```python
fig, axs = plt.subplots(2, 2, constrained_layout=True)
for ax in axs.flat:
    example_plot(ax)
```
