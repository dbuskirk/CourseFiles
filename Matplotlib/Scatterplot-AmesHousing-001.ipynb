```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path= "../data/"
```

```python
dfHousing = pd.read_csv(path + 'AmesHousing.csv')
```

```python
# dfHousing.info()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = dfHousing['Gr Liv Area'], y = dfHousing['SalePrice'])
plt.xlabel("Living Area Above Ground")
plt.ylabel("House Price")

plt.show()
```

```python
fig, ax = plt.subplots(2, figsize=(10, 6))
ax[0].scatter(x = dfHousing['Gr Liv Area'], y = dfHousing['SalePrice'])
ax[0].set_xlabel("Living Area Above Ground")
ax[0].set_ylabel("House Price")

ax[1].scatter(x = dfHousing['Overall Qual'], y = dfHousing['SalePrice'])
ax[1].set_xlabel("Overall Quality")
ax[1].set_ylabel("House Price")

plt.show()
```

Observe that in the following multiple plot the vertical axis label for the second plot spills over into the first plot. This sort of thing is a common occurrence and might be easy-to-fix with <b>constrained_label=True</b>.

```python
fig, ax = plt.subplots(2,2, figsize=(10,10))
ax[0,1].scatter(x = dfHousing['Gr Liv Area'], y = dfHousing['SalePrice'])
ax[0,1].set_xlabel("Living Area Above Ground")
ax[0,1].set_ylabel("House Price")

plt.show()
```

Here we apply <b>constrained_label=True</b>.

```python
fig, ax = plt.subplots(2,2, figsize=(10,10), constrained_layout=True)
ax[0,1].scatter(x = dfHousing['Gr Liv Area'], y = dfHousing['SalePrice'])
ax[0,1].set_xlabel("Living Area Above Ground")
ax[0,1].set_ylabel("House Price")

plt.show()
```
